**Heatmap of Indian States and Union Teritories taking important parameters as a criterion**

```python
plt.figure(figsize=(12,10))
plt.style.use('fivethirtyeight')
sns.set_context('notebook',font_scale=1)
sns.set_style('whitegrid')
st= sc.pivot_table(values='Total',index='State',columns='Type_code')
sns.heatmap(st,linewidths=3,cmap='Set2_r')
```

```python
plt.figure(figsize=(12,10))
plt.style.use('fivethirtyeight')
sns.set_context('notebook',font_scale=1)
sns.set_style('whitegrid')
st= sc.pivot_table(values='Total',index='State',columns='Type')
sns.heatmap(st, linewidths=2,cmap= 'RdYlGn_r')
```

*From the above two tables and data given, the conclusion taken is most people have used hanging as a means to suicide, and the second means is by consuming insecticides or poison. The cause of suicide is mostly family problems and somewhat due to diseases. Most of the cases are from married and divorce or widow peoples. Women who are housewives, people engaged in farming/agricultural activities, children in primary, middle school, or no education suicided the most. The number of suicides increased in 2004 and from 2008 onwards took a massive increment. Males and the age group between 15-45 were involved in most suicides. Analysis of statewise and union territories is hard to interpret as the data is almost the same for all states and union territories for five parameters.
